# Scikit-learn Design

```python
# Primarily there are three types of objects
#1. Estimators-> It estimates some parameters based on a dataset.
#    Eg. Imputer(that we used above)
#    It has a fit method and transform method.
#    Fit Method - Fits the dataset and calculations internal parameters.
#2. Transformers-> Transform method takes input and returns output based on the learnings from fit(). It has a convience function called fit_transform() which fits and then transforms.
#3. Predictors-> LinearRegression model is an example of a predictor. fit() and predict() are two common functions. It also give score() function which will evaluate the predictions.
```

# Feature Scaling

```python
#To have the different column values of the dataset in same scaling
#For eg. CHAM is ranging from 0 to 1 whereas RM is ranging from 1 to 8 and to have them on same scale we do this 
#Primarily there are two types of feature scaling methods:
#    1) Min-Max scaling(Normalization)
#        (value-min)/(max-min)
#        Sklearn provides a class called MinMaxScalar for this
        
#    2) Standardization
#        (value-mean)/STD 
#        Sklearn provides a class called Standard Scalar for this
        
#We will use standardization because min and max can change significantly if there is a correction but STD and mean usually doesn't change that much.
```

# Creating a Pipeline


```python
housing.describe()
```

```python
# Pipelining will do the imputer function and Scaling
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    #...add as many as we want in pipeline....
    ('std_scaler', StandardScaler())
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing)
# The resultant is a numpy array which we need as input to Predictors
```

# Using better evaluation technique- Cross Validation

```python
# Explanation of technique that we will be using: We will divide the testing data into for example 10 groups.
# -> Train the model on 2 to 10 groups and test with group 1.
# -> Then train the model on 1 to 10 groups except 2 and test with group 2.
# -> Then train the model on 1 to 10 groups except 3 and test with group 3.
# and so on
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, housing_num_tr, housing_labels, scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)# As scores are negative
```
